```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Load data
file_path = Path("Resources/sample_checkpoint.csv")
df = pd.read_csv(file_path)
df.head(10)
```

```python
df.count()
```

```python
# value counts of lien status column
df["lien_status"].value_counts()
```

```python
# Change values in lien status to reflect categorical variables
df['lien_status'] = df['lien_status'].replace([1,2,3,4],['Secured by First Lien','Secured by a Subordinate Loan','Not Secured by Lien','Not Applicable (Purchased Loan)'])
df["lien_status"].value_counts()
```

```python
df.dtypes
```

```python
df5 = df.drop(columns=["state_code"])
df5.head(10)
```

```python
final_df = df5.groupby("action_taken", group_keys=False).apply(lambda x: x.sample(frac=0.01))
```

```python
final_df.head(10)
```

```python
final_df.count()
```

```python
final_df.info()
```

```python
save_path = "Resources/final_sample.csv"
final_df.to_csv(save_path, index=False)
```
